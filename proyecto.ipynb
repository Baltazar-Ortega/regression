# Regresion Lineal

<p>
La regresion lineal busca encontrar una linea recta que se adapte lo mejor posible a los datos
</p>
<p>
Este modelo "aprenderá" de datos de entrenamiento, y podrá hacer predicciones de datos que no conozca

## Datos del problema

```julia
h = 0.01 # pequeño cambio para la derivada
lr = 0.001 # learning rate

## DATOS ##
# Años de experiencia
x = [3, 5, 6]
# Salario en miles
y = [10, 14, 17]
# Cantidad de registros
n = 3

# Inicializacion de parametros
w = 2
bias = 3
v = [w, bias]
```

```julia
xLinea = [0, 6]
yLinea = [3, 15]

using Plots
plot(xLinea, yLinea, legend=false, xlabel="tiempo", ylabel="Salario (miles)")
scatter!(x, y, title = "Mi primer modelo")
```

```julia
function suma_cuadrados_mse(y, x, v)
    suma = 0
    for i = 1:length(x)
        suma += ( y[i] - ( (x[i] * v[1]) + v[2] ) ) ^ 2
    end
    return suma
end

function mse(x, y, n, vector_parametros)
    suma = suma_cuadrados_mse(y, x, vector_parametros)
    return suma / n
end

function prediccion(x, v)
    return [( v[1] * x_i ) + v[2] for x_i in x]
end
```

```julia
mse_actual = mse(x, y, n, v)
println("mse actual: ", mse_actual)
```

```julia
function sumar_h(i, j)
    if i == j
        return h
    else
        return 0
    end
end
```

```julia
function parcial(i, v)
    v2 = Float64[]
    for (j, v_i) in enumerate(v)
        suma_h_condicionado = v_i + sumar_h(i, j)
        append!(v2,  suma_h_condicionado )
    end
    return ( mse(x, y, n, v2) - mse(x, y, n, v) ) / h
end

function gradiente(v)
    return [parcial(i, v) for i in 1:length(v) ]
end
```

```julia
gradiente(v)
```

```julia
function distancia(v, w)
    return sqrt( sum( (x - y) ^ 2 for (x, y) in zip(v, w) ) )
end

function paso(v, gradiente)
    # se multiplica por learning rate para el cambio no sea brusco y pueda provocar errores
    return [param - (lr * grad_i) for (param, grad_i) in zip(v, gradiente) ]
end
```

```julia
convergencia = 0.00001
i = 1
while true
    grad = gradiente(v)
    v2 = paso(v, grad) # hacer un paso encontra de la derivada (ir hacia el minimo global)
    if distancia(v2, v) < convergencia
        println("\nSe alcanza convergencia luego de ",i, " iteraciones")
        println("\nValor mínimo de la función en:\n v = ", v)
        break
    end
    if i > 10000
        println("Salida luego de ",i, " iteraciones")
        break
    end
    println("iteración:", i, "\n","v =", v)
    global v = v2
    global i = i + 1
end
```

```julia
# Prediccion
x = [4]  # Este dato no se encuentra en los datos de entrenamiento
prediccion(x, v) # Salario de programador con 4 años de experiencia
```

## Ridge Regression

```julia
# Datos de entrenamiento
x =  [0.5, 1.7]
y = [1.05, 2.61]
n = 2
# pendiente y bias de la recta con error mse = 0
v = [1.3, 0.4]
```

```julia
plot([0.5, 4], [1.05, 5.6], color="red")
scatter!(x, y, legend=false, color="red")
```

```julia
# lambda 1-200
function mse_ridge(x, y, n, vector_parametros, lambda)
    mse = ( suma_cuadrados_mse(y, x, vector_parametros) ) / n
    return mse + ( lambda * (vector_parametros[1] ^ 2) )
end
```

```julia
# Ahora hacemos un modelo para esos datos, pero con ridge
lambda = 1
linea_pegada = mse_ridge(x, y, n, v, lambda) # con la linea pegada a los puntos
print("\nLinea pegada: ", linea_pegada)

v_sep = [0.8, 0.9]
linea_separada_poco = mse_ridge(x, y, n, v_sep, lambda)
print("\nLinea separada un poco: ", linea_separada_poco)
```

```julia
x_sep = [0.5, 1.6]
y_sep = [1.3, 2.18]
plot(x_sep, y_sep, color="blue")
plot!(x, y, color="red")
scatter!(x, y, legend=false, color="red")
```

```julia
# Datos de test
x_test = [0.9, 2.1, 2.8, 3.1, 3.8, 4.4]
y_test = [1.8, 1.4, 2.6, 3.7, 3.1, 3.6]
```

```julia
scatter(x, y, legend=false, color="red")
scatter!(x_test, y_test, legend=false, color="green")
plot!([0.5, 4], [1.3, 4.1], color="blue")
plot!([0.5, 2.5], [1.05, 3.65], color="red")
```

## Lasso Regression

```julia
# lambda 1-200
function mse_lasso(x, y, n, vector_parametros, lambda)
    mse = ( suma_cuadrados_mse(y, x, vector_parametros) ) / n
    return mse + ( lambda * abs(vector_parametros[1]) )
end
```

```julia
# Datos de entrenamiento
x =  [0.5, 1.7]
y = [1.05, 2.61]
n = 2
# pendiente y bias de la recta con error mse = 0
v = [1.3, 0.4]
```

```julia
# Ahora hacemos un modelo para esos datos, pero con lasso
lambda = 1
linea_pegada = mse_lasso(x, y, n, v, lambda) # con la linea pegada a los puntos
print("\nLinea pegada: ", linea_pegada)

v_sep = [0.7, 1.1]
linea_separada_poco = mse_lasso(x, y, n, v_sep, lambda)
print("\nLinea separada un poco: ", linea_separada_poco)
```

```julia
x_sep = [0.5, 1.6]
y_sep = [1.25, 2.46]
plot(x_sep, y_sep, color="brown")
plot!(x, y, color="red")
scatter!(x, y, legend=false, color="red")
```

## Elastic Net

```julia
# lambda1 1-200, lambda2 1-200
function mse_elastic_net(x, y, n, vector_parametros, lambda_1, lambda_2)
    mse = ( suma_cuadrados_mse(y, x, vector_parametros) ) / n
    ridge = lambda_1 * (vector_parametros[1] ^ 2) 
    lasso = lambda_2 * abs(vector_parametros[1])
    return mse + ridge + lasso
end
```

```julia
# Datos de entrenamiento
x =  [0.5, 1.7]
y = [1.05, 2.61]
n = 2
# pendiente y bias de la recta con error mse = 0
v = [1.3, 0.4]
```

```julia
# Ahora hacemos un modelo para esos datos, pero con lasso
lambda_1 = 1
lambda_2 = 1
linea_pegada = mse_elastic_net(x, y, n, v, lambda_1, lambda_2) # con la linea pegada a los puntos
print("\nLinea pegada: ", linea_pegada)

v_sep = [0.9, 0.8]
linea_separada_poco = mse_elastic_net(x, y, n, v_sep, lambda_1, lambda_2)
print("\nLinea separada un poco: ", linea_separada_poco)
```

```julia
x_sep = [0.5, 1.6]
y_sep = [1.25, 2.24]
plot(x_sep, y_sep, color="grey")
plot!(x, y, color="red")
scatter!(x, y, legend=false, color="red")
```

## Regresion Lineal con videojuegos

### Datos

```julia
# Datos
x = features
y = labels
```

### Funciones de evaluacion

```julia
function prediccion(array, parametros)
    suma_pond = 0
    for i = 1:length(array)
        suma_pond += array[i] * parametros[i]
    end
    return suma_pond + parametros[end]
end

# x es una matriz (dataframe)
function mse(x, y, n, parametros)
    suma = 0
    for i = 1:size(x, 1)
        y_pred = prediccion(x[i, :], parametros)
        suma += ( y[i] - y_pred ) ^ 2
    end
    return suma / n
end
```

### Parámetros

```julia
h = 0.01 # pequeño cambio para la derivada (hiperparametro)
lr = 0.001 # learning rate (el mejor resultado se obtuvo con 0.0001, pero tarda mucho)

# Cantidad de registros
n = size(features, 1)

# Inicializacion de parametros
w1 = 2
w2 = 3
w3 = 4
w4 = 5
bias = 4
parametros = [w1, w2, w3, w4, bias] 
```

```julia
print("mse actual: ", mse(x, y, n, parametros))
```

### Funciones para algoritmo

```julia
function sumar_h(i, j)
    if i == j
        return h
    else
        return 0
    end
end

function parcial(i, v)
    v2 = Float64[]
    for (j, v_i) in enumerate(v)
        suma_h_condicionado = v_i + sumar_h(i, j)
        append!(v2,  suma_h_condicionado )
    end
    derivada_parcial = ( mse(x, y, n, v2) - mse(x, y, n, v) ) / h
    return derivada_parcial
end

function gradiente(v)
    return [parcial(i, v) for i in 1:length(v) ]
end

function distancia(v, w)
    return sqrt( sum( (x - y) ^ 2 for (x, y) in zip(v, w) ) )
end

function paso(v, gradiente)
    # se multiplica por learning rate para el cambio no sea brusco y pueda provocar errores
    return [param - (lr * grad_i) for (param, grad_i) in zip(v, gradiente) ]
end
```

```julia
convergencia = 0.00000001
i = 1
while true
    grad = gradiente(parametros)
    parametros2 = paso(parametros, grad) # hacer un paso encontra de la derivada (ir hacia el minimo global)
    if distancia(parametros2, parametros) < convergencia
        println("\nSe alcanza convergencia luego de ",i, " iteraciones")
        println("\nValor mínimo de la función en:\n parametros = ", parametros)
        break
    end
    if i > 10000
        println("Salida luego de ",i, " iteraciones")
        break
    end
    println("iteración:", i, "\n","parametros =", parametros)
    global parametros = parametros2
    global i = i + 1
end
```

# Metodos de regularizacion

<p>
Muchas veces puede ser que nuestro modelo de regresion lineal haga overfitting de los datos de 
entrenamiento. Esto significa, que "memorizó" en lugar de "generalizar". 
</p>
<p>
Existen metodos de regularizacion que se encargan de evitar este fenomeno. <br>
    Solo se distinguen de la regresión lineal en la <strong>funcion de perdida.</strong> 
</p>
<p>
Ademas, utilizan un hiperparametro (lambda) que puede tener un valor entre 1-inf. <br>
En la vida real se evalua el modelo con cientos de valores para lambda y se escoge el mejor.<br>
Por terminos de alcance de proyecto y evitar tomar mucho tiempo, se dejó el valor de lambda en 1.
</p>
<ul>
    <li><strong>Ridge: </strong>reduce el valor de los features poco relevantes. <br>Ej. Puede ser que las calificaciones de los criticos no tengan mucha importancia.</li>
    <li><strong>Lasso: </strong>elimina los features que no influyen mucho en el calculo</li>
    <li><strong>Elastic Net: </strong>Es una combinacion de Ridge y Lasso</li>
<ul>
